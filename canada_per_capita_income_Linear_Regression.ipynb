```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model as lm
Read_data = pd.read_csv('canada_per_capita_income.csv',sep=';')
Read_data.columns = Read_data.columns.str.strip()

print(Read_data.head())

print(Read_data.describe())

```

Scatter the data
رسم البيانات

```python
plt.scatter(Read_data['year'], Read_data['capita income (US$)'], color='blue', marker='+')
plt.xlabel('year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Over Years')
plt.show()
```

train the model (linear regression model  using the avalible data point    )
تدريب المودل

```python
lmodel=lm.LinearRegression()
lmodel.fit(Read_data[['year']],Read_data[['capita income (US$)']])
```

We have train the linear Regression model on The Date <**> now we want to predicate

```python
print('Prediction for 2011 is : ',lmodel.predict([[2011]]))
```

Income = Slope × Year + Intercept
+ 
بعد تدريب النموذج، من المهم اعرف المعادلة اللي النموذج تعلّمها
+ 
احسب جوده النموذج 

```python
print("Slope (coefficient):", lmodel.coef_)
print("Intercept:", lmodel.intercept_)

print("Model R² score:", lmodel.score(Read_data[['year']], Read_data[['capita income (US$)']]))
```

```python
plt.scatter(Read_data['year'], Read_data['capita income (US$)'], color='blue', marker='+')
plt.xlabel('year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Over Years')
plt.plot(Read_data['year'], lmodel.predict(Read_data[['year']]), color='red')
plt.grid(True)
plt.show()
```
